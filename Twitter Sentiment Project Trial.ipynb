## Cleaning of Tweets

```python
import nltk
from nltk.corpus import stopwords
import string
```

```python
stop=stopwords.words('english')
punc=list(string.punctuation)
stop = punc+stop
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def clean(text):
    words=word_tokenize(text)
    clean_words=[]
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            clean_words.append(clean_word)
    clean_text= " ".join(clean_words)
    return clean_text
            
            
            
        
```

```python
tweets = [clean(tweet) for tweet in tweets]
tweets
```

```python
from sklearn.model_selection import train_test_split

tweets_train,tweets_test,sent_train,sent_test = train_test_split(tweets,sentiment)
```

```python

from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer(max_features=5000,max_df=0.8,ngram_range = (1,2))

X_train = count_vec.fit_transform(tweets_train)
Y_train = sent_train
```

### Trying SVC

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score,KFold,GridSearchCV
clf_1 = SVC(C=10,gamma = 0.01)

clf_1.fit(X_train,Y_train)
```

```python
X_test = count_vec.transform(tweets_test)
Y_test = sent_test
```

```python
Y_Pred=clf_1.predict(X_test)
clf_1.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(Y_test,Y_Pred))
print(confusion_matrix(Y_test,Y_Pred))
```

### Trying RandomForest

```python
from sklearn import ensemble
clf_2 = ensemble.RandomForestClassifier()
```

```python
clf_2.fit(X_train,Y_train)
```

```python
Y_Pred=clf_2.predict(X_test)
clf_2.score(X_test,Y_test)
```
